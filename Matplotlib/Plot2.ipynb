```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = np.linspace(0,10,20)
```

```python
a
```

```python
b = a ** 2
```

```python
(myFigure, myAxis) = plt.subplots()
myAxis.plot(a,b,"g")
myAxis.plot(b,a,"b")
```
