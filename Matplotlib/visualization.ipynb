```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,20,20)
```

```python
b = a ** 3
```

```python
plt.plot(a,b,"g")
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.8,0.8,0.3,0.3])
figureAxes.plot(a,b,"g")
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik İsmi")
```

```python
fgr = plt.figure()
eksen1 = fgr.add_axes([0.1,0.1,0.7,0.7])
eksen2 = fgr.add_axes([0.2,0.4,0.3,0.3])

eksen1.plot(a,b,"g")
eksen2.plot(b,a,"b")
```

```python
newFigure = plt.figure(dpi=100)
newAxis = newFigure.add_axes([0.1,0.1,0.9,0.9])
newAxis.plot(a,a ** 2,label="Uzeri ** 2")
newAxis.plot(a,a ** 3,label="Uzeri ** 3")
newAxis.legend(loc=2)
```
